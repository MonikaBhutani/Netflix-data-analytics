```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset

```python
df= pd.read_csv("BigBasket Products.csv")
```

```python
# Step 4: Dataset Description
df.describe(include='all')
```

```python
# Step 5: DataFrame Info
df.info()
```

```python
# Step 6: Top and Least Sold Products (assumption: based on rating)
print("\nTop 5 Products based on rating:")
df.sort_values(by='rating', ascending=False).head()
```

```python
top_rated = df.sort_values(by='rating', ascending=False).head(5)
least_rated = df.sort_values(by='rating').head(5)

```

```python
# Step 7: Measuring Discount
# Discount = market_price - sale_price
df['discount'] = df['market_price'] - df['sale_price']
print("\nTop 5 Products with Highest Discounts:")
df[['product', 'sale_price', 'market_price', 'discount']].sort_values(by='discount', ascending=False).head()
```

```python
# Step 8: Missing Values
print("\nMissing Values in Dataset:")
print(df.isnull().sum())
```

```python
# Step 10: Visualizations
plt.figure(figsize=(10,6))
sns.histplot(df['sale_price'], kde=True)
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='category', y='sale_price', data=df)
plt.xticks(rotation=45)
plt.title('Sale Price by Category')
plt.show()

```

```python
plt.figure(figsize=(10,6))
top_brands = df['brand'].value_counts().head(10)
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title('Top 10 Brands')
plt.xticks(rotation=45)
plt.ylabel('Product Count')
plt.show()
```
